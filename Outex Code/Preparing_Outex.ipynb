```python
from PIL import Image
from glob import glob
```

```python
import os
```

```python
mode = 0o666
```

```python
for i in range(1,25):
  os.mkdir('Class'+str(i),mode)
  path=paths[-20+i*20:20*i]+paths[480+(-20+i*20):480+i*20]
  k=1
  for image_path in path:
    rawData=open('images/'+str(image_path),'rb').read()
    imgSize = (128,128)# the image size
    img = Image.frombytes('L', imgSize, rawData)
    img.save('Class'+str(i)+'/s'+str(k)+".png")
    k=k+1
    #i=i+1# can give any format you like .png
#k=k+1
```
